### label encoding

```python
### 마스크 여부, 성별, 나이를 mapping할 클래스를 생성합니다.

class MaskLabels(int, Enum):
    MASK = 0
    INCORRECT = 1
    NORMAL = 2


class GenderLabels(int, Enum):
    MALE = 0
    FEMALE = 1

    @classmethod
    def from_str(cls, value: str) -> int:
        value = value.lower()
        if value == "male":
            return cls.MALE
        elif value == "female":
            return cls.FEMALE
        else:
            raise ValueError(f"Gender value should be either 'male' or 'female', {value}")


class AgeLabels(int, Enum):
    YOUNG = 0
    MIDDLE = 1
    OLD = 2

    @classmethod
    def from_number(cls, value: str) -> int:
        try:
            value = int(value)
        except Exception:
            raise ValueError(f"Age value should be numeric, {value}")

        if value < 30:
            return cls.YOUNG
        elif value < 60:
            return cls.MIDDLE
        else:
            return cls.OLD
```

```python
num_classes = 3 * 2 * 3

_file_names = {
    "mask1": MaskLabels.MASK,
    "mask2": MaskLabels.MASK,
    "mask3": MaskLabels.MASK,
    "mask4": MaskLabels.MASK,
    "mask5": MaskLabels.MASK,
    "incorrect_mask": MaskLabels.INCORRECT,
    "normal": MaskLabels.NORMAL
}

image_paths = []
mask_labels = []
gender_labels = []
age_labels = []
```

```python
def setup(img_dir):
    """
    image의 경로와 각 이미지들의 label을 계산하여 저장해두는 함수입니다.
    """
    profiles = os.listdir(img_dir)
    for profile in profiles:
        if profile.startswith("."):  # "." 로 시작하는 파일은 무시합니다
            continue

        img_folder = os.path.join(img_dir, profile)
        for file_name in os.listdir(img_folder):
            fn, ext = os.path.splitext(file_name)
            if fn not in _file_names:  # "." 로 시작하는 파일 및 invalid 한 파일들은 무시합니다
                continue

            img_path = os.path.join(img_dir, profile, file_name)  # (resized_data, 000004_male_Asian_54, mask1.jpg)
            mask_label = _file_names[fn]

            id, gender, race, age = profile.split("_")
            gender_label = GenderLabels.from_str(gender)
            age_label = AgeLabels.from_number(age)

            image_paths.append(img_path)
            mask_labels.append(mask_label)
            gender_labels.append(gender_label)
            age_labels.append(age_label)
```

```python
img_dir = '/opt/ml/input/data/train/images'
setup(img_dir)
```

```python
image_list = []
multi_label_list = []
multi_name_list = []
multi_label_name_dict = {}

for index in range(len(image_paths)):
    image_path = image_paths[index]
#     image = Image.open(image_path)
    image = cv2.imread(image_path)[:,:,::-1]
    
    mask_label = mask_labels[index]
    gender_label = gender_labels[index]
    age_label = age_labels[index]
    multi_class_name = f'{mask_label.name}_{gender_label.name}_{age_label.name}'
    multi_class_label = mask_label * 6 + gender_label * 3 + age_label
    
    image_list.append(image)
    multi_name_list.append(multi_class_name)
    multi_label_list.append(multi_class_label)
```

```python
class MyCrop:
    """Rotate by one of the given angles."""

    def __init__(self,top=512-473, left=30, height=420, width=384-60):
        self.top = top
        self.left = left
        self.height = height
        self.width = width

    def __call__(self, x):
        return TF.crop(x, self.top, self.left, self.height, self.width)
```
